# Qualitative Data

```python
# Qualitative data example (Text and categories)
# Creating lists with qualitative data
product=["apple","banana","cherry","apple","banana"]
reviews=["good","average","excellent","good","poor"]



# Displaying the qualitative data list
print("Qualitative Data:")
print("Productive:",product)
print("Reviews:",reviews)
```

```python
# Demonstrating data types
print("\nData types of Qualitative data:")
print("products:",type(product))
print("reviews:",type(reviews))
```

```python
# Demonstrating data types
print("Data types of Qualitative data:")
print("products:",type(product[0]))
print("reviews:",type(reviews[0]))
```

# Quantitative Data

```python
# Quantitative data example (integer and floats)
# Creating lists with quantitative data
product_ids=[101,102,103,104,105]
prices=[0.99,0.5,1.99,0.99,0.5]
quantities=[10,20,15,10,5]


# Displaying the quantitative data lists
print("quantitative data:")
print("product IDs:",product_ids)
print("prices:",prices)
print("quantitative:",quantities)
```

```python
# Demonstrating data types
print("Data types of quantitative:")
print("product IDs:",type(product_ids))
print("prices:",type(prices))
print("quatities:",type(quantities))
```

```python
# Demonstrating data types
print("Data types of quantitative:")
print("product IDs:",type(product_ids[0]))
print("prices:",type(prices[0]))
print("quatities:",type(quantities[0]))
```
