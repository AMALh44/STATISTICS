```python
# Importingg necessary libraries
import numpy as np
import pandas as pd

from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
# Example dataset:Actual vs predicted values
actual=np.array([3,-0.5,2,7])
predicted=np.array([2.5,0.0,2,9])
```

```python
# Creating a  DataFrame
df=pd.DataFrame({
    "actual":actual,
    "predicted":predicted
})
print(df)
```

# The mean absolute error-(MAE)

```python
# Calculating mean absolute error (MAE) using function
mae=mean_absolute_error(actual,predicted)
print("Mean Absolute Error (MAE):",mae)
```

# The mean squared error (MSE)

```python
# Calculating the mean squared error (MSE) using function
mse=mean_squared_error(actual,predicted)
print("Mean squared error (MSE):",mse)
```
