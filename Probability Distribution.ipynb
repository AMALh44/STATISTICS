```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint
```

# Uniform Distrbution

A Uniform distribution has constant probability within a given range.

The Uniform distribution is a continuous distribution in which all outcomes in an interval [a,b] are equally likely.

It is defined by parameters:

* a:lower bound
* b:Upper bound

Example:Modeling the probability of a random point along a line segment.

```python
from scipy.stats import uniform

# parameters
a,b = 0,10
x_values = np.linspace(a, b, 1000)

# plot PDF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x_values, uniform.pdf(x_values, a, b-a), color='blue')
plt.title('Uniform Distribution PDF (a = 0, b = 10)')
plt.xlabel('x')
plt.ylabel('PDF')

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x_values, uniform.cdf(x_values, a, b-a), color='green')
plt.title('Uniform Distribution CDF (a = 0, b = 10)')
plt.xlabel('x')
plt.ylabel('CDF')

plt.tight_layout()
```

```python
# Parameters for Discrete Uniform Distribution
low,high = 1,6 # Inclusive range from 1 to 6 (like a fair die)
x_values = np.arange(low, high + 1)

plt.figure(figsize=(12, 6))

# Plot PMF
plt.subplot(1, 2, 1)
plt.stem(x_values, randint.pmf(x_values, low, high + 1), basefmt=' ')
plt.title('Discrete Uniform Distributionn PMF (low = 1, high = 6)')
plt.xlabel('x')
plt.ylabel('PMF')


# Plot CDF
plt.subplot(1, 2, 2)
plt.step(x_values, randint.cdf(x_values, low, high + 1), color='green')
plt.title('Discrete Uniform Distributionn CDF (low = 1, high = 6)')
plt.xlabel('x')
plt.ylabel('CDF')
```

# Bernoulli Distribution

The Bernoulli Distribution models a single experiment with two possible outcomes: success(With probability p) and failure (with probability 1 - p).

It is a discrete distribution and defined by the parameter:

* p: Probability of success

Example:Modeling a single coin flip where heads is asuccess and tails is a failure.

```python
from scipy.stats import bernoulli

# Parameters
p = 0.5
x_values = np.array([0, 1])

plt.figure(figsize=(12, 6))
# Plot PMF
plt.subplot(1, 2, 1)
plt.stem(x_values, bernoulli.pmf(x_values, p), basefmt=" ")
plt.title("Bernoulli Distribution PMF (p = 0.5)")
plt.xlabel('x')
plt.ylabel('PMF')

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(x_values, bernoulli.cdf(x_values, p), where='post', color='green')
plt.title("Bernoulli Distribution CDF (p = 0.2)")
plt.xlabel('x')
plt.ylabel('CDF')

plt.show()
```

# Binomial Distribution

The Binomial Distribution Models the number of successes in afixed number of independent trials. Each with the same probability of success.

Example:Modeling the number of heads in 10 coin flips.

It is dicrete and defined by parameters:
* n:Number of trials
* p:Probability of success in each trial

```python
from scipy.stats import binom

# Parameters for Binomial Distribution
n, p = 10, 0.5
x = np.arange(0, n+1)

plt.figure(figsize=(12, 6))
# plot PMF
plt.subplot(1,2,1)
plt.bar(x, binom.pmf(x, n, p), color='skyblue')
plt.title('Binomial Distribution  PMF')
plt.xlabel('x')
plt.ylabel('p(X=x)')

# plot CDF
plt.subplot(1,2,2)
plt.step(x, binom.cdf(x, n, p), color='orange')
plt.title('Binomial Distribution  CDF')
plt.xlabel('x')
plt.ylabel('p(X  x)')

plt.show()
```

# Normal Distribution

> The Normal Distribution, also known as Gaussian distribution, is a continous distribution defined by its symmetric, bell-shaped curve centered around the mean.

> It is defined by parameters:

>  - **mu**:Mean of the distribution
>  - **sigma**:Standard deviation, which determines the spread of the distribution

>> Speacial case: The Standard normal distribution has 'mu = 0' and 'sigma = 1'

> Example: Modeling heights, test scores, or measurement errors

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Parameters for normal Distribution
mu = 1      # Mean
sigma = 3   # Standard deviation
x = np.linspace(-10, 10, 100)

plt.figure(figsize=(12, 6))
# plot PDF
plt.subplot(1,2,1)
plt.plot(x, norm.pdf(x, mu, sigma), color='blue', label='PDF')
plt.title("Normal Distribution PDF")
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.legend()


# CDF
plt.subplot(1, 2, 2)
plt.plot(x, norm.cdf(x, mu, sigma), color='g', label='CDF')
plt.title("Normal Distribution CDF")
plt.xlabel('x')
plt.ylabel('F(x  x)')
plt.grid(True)
plt.legend()

plt.show()
```

# Poisson Distribution

The poisson Distribution models the number of events occurrinng in a fixed interval of time or space when these events occur independently and with a constant average rate (lambda or mu).

it is discrete and defined by parameter:

* mu (lambda): Average number of events per interval.

Example: Modeling the number of emails received per hour.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Parameters for poisson Distribution
mu = 3      # Mean
x = np.arange(0, 15)

plt.figure(figsize=(12, 6))
# plot PMF
plt.subplot(1,2,1)
plt.bar(x, poisson.pmf(x, mu), color='blue')
plt.title("poisson Distribution PMF")
plt.xlabel('x')
plt.ylabel('p(X=x)')


# CDF
plt.subplot(1, 2, 2)
plt.step(x, poisson.cdf(x, mu), color='purple')
plt.title("poisson Distribution CDF")
plt.xlabel('x')
plt.ylabel('p(x  x)')

plt.show()
```

# Exponential Distribution

The Exponenetial Distribution models the time between independent events that accur at a constant rate.

It is a continuous distribution and is defined by parametter:

* lambda_:Rate parameter, which is the inverse of the mean

example: Modeling the time between arrivals at a service station.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon

# Parameters for Exponenetial Distribution
lambda_ = 1
x = np.linspace(0, 5, 100)

plt.figure(figsize=(12, 6))
# plot PdF
plt.subplot(1,2,1)
plt.plot(x, expon.pdf(x, scale=1 / lambda_), color='cyan')
plt.title("Exponenetial Distribution PDF")
plt.xlabel('x')
plt.ylabel('f(x)')


# CDF
plt.subplot(1, 2, 2)
plt.plot(x, expon.cdf(x, scale=1/lambda_), color='magenta')
plt.title("Exponenetial Distribution CDF")
plt.xlabel('x')
plt.ylabel('p(Xx)')

plt.show()
```

# Gamma Distribution

The Gamma Distribution is a continuous distribution defined for positive values.

It is often used to model waiting times and is defined by parameters:

* k:Shape parameter
* theta:Scale parameter (or 1/ rate)

Example:Modeling the time until a specified number of events occur, such as time until a machine fails.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gamma

# Parameters for Exponenetial Distribution
k = 2
theta = 2
x = np.linspace(0, 20, 100)

plt.figure(figsize=(12, 6))
# plot PDF
plt.subplot(1,2,1)
plt.plot(x, gamma.pdf(x, k, scale=theta), color='purple', label='PDF')
plt.title("gamma Distribution PDF")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# CDF
plt.subplot(1, 2, 2)
plt.plot(x, gamma.cdf(x, k, scale=theta), color='purple', label='CDF')
plt.title("gamma Distribution CDF")
plt.xlabel('x')
plt.ylabel('p(Xx)')
plt.grid(True)
plt.legend()

plt.show()
```
