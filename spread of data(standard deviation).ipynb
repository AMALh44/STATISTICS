```python
# import necessary libraries
import  numpy as np
import pandas as pd
import statistics as stats
```

# Data

```python
# sample data
data=[5,7,9,10,12]
```

the sample mean and population mean are same in the example

```python
# step 1:calculate the mean
mean=np.mean(data)
print("population mean = sample mean =",mean)
```

# Calculating variance and standard deviation using formula

```python
# step 2:
data=[5,7,9,10,12]
mean=np.mean(data)
differences =[x-mean for x in data]
squared_differences =[d**2 for d in differences]
```

```python
# step 3:
data={
    "data":[5,7,9,10,12],
    "---differences from mean":[x-mean for x in data],
    "---squared_differences":[d**2 for d in differences]
}
# converting the data into dataframe
a=pd.DataFrame(data)
# display the dataset
print("dataset")
print(a)
```

```python
# step 4:calculate the population variance
pop_var=sum(squared_differences)/len(data)
print("variance:",pop_var)
```

```python
# step 5:caculate the population standard deviation
pop_std=np.sqrt(pop_var)
print("standard deviation",pop_std)
```

# Calculating variance and standard deviation using statistics module

## The sample variance

```python
data=[5,7,9,10,12]
var=stats.variance(data)
print("the sample variance is:",var)
```

```python
stdev=stats.stdev(data)
print("Standard Deviantion is:",stdev)
```

## The population variance

```python
p_var=stats.pvariance(data)
print("the population variance is:",p_var)
```

```python
p_stdev=stats.pstdev(data)
print("the population standard deviation is:",p_stdev)
```
