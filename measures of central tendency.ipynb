```python
# import the necessary libraries
import numpy as np
import statistics as stats
```

```python
# data for examples
data=[10,15,20,25,30,35,40,45,50,55,60]
```

# MEAN


```python
# sum of all values divided by the number of value
(10+15+20+25+30+35+40+45+50+55+60)/11
```

```python
# mean example
mean_value=np.mean(data)
print("mean(average):")
print("data:",data)
print("mean is:",mean_value)
```

```python
# mean using statistics module
mean_value=stats.mean(data)
print("mean(average):")
print("data:",data)
print("mean is:",mean_value)
```

```python
print(stats.mean([10,15,20,25,30,35,40,45,50,55,60]))
```

```python
print(stats.mean([1.3,2.5,6.5,4.5,8.9,3.6,7.8,13.12]))
```

# Median

```python
# meadian example
meadian_value=np.median(data)
print("median:")
print("data:",data)
print("median is:",meadian_value)
```

```python
# median using statistics
meadian_value=stats.median(data)
print("median:")
print("data:",data)
print("median is:",meadian_value)
```

if the number of data values is odd,it return the exact middle value.if the number of value even,it return the average of two middle values.

```python
# the number of data values is odd
print(stats.median([1,3,5,7,9,11,13,15,17]))
```

```python
# the number of dat value is even
print(stats.median([2,4,6,8,10,12,14,16]))
```

```python
print(stats.median([-11,5.5,-3.4,7.1,-9,22]))
```

## median high


calculate the high median of the given data,if the number of data value is even,it returns the larger of the two middle values.

```python
print(stats.median_high([1,2,3,4,5,6,7,8]))
```

```python
print(stats.median_high([-11,5.5,-3.4,7.1,-9,22]))
# -11,-9,-3.4,5.5,7.1,22
```

## median low


```python
print(stats.median_low([1,3,5,7,9,23]))
```

```python
print(stats.median_low([-11,5.5,-3.4,7.1,-9,21]))
```

# Mode


if a value occur with the same frequency,statistics.mode() returns thhe first value in the order of the dataset,which is 10 in the case

```python
# mode using statistics module
mode_value=stats.mode(data)
print("mode is:",mode_value)
# this data has no repeating value
```

```python
# mode example
a=[1,2,2,3,3,3,4,4,4,4,5]
x=stats.mode(a)
print("mode is:",x)
```

```python
print(stats.mode(["yes","no","yes","no","no","yes","no"]))
```

```python
print(stats.mode(["red","green","blue","red"]))
```
