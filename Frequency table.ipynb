```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create sample data
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'B', 'C', 'A', 'C', 'B', 'A', 'C'],
    'Scores': [85, 90, 88, 75, 95, 88, 90, 78, 88, 80, 90, 78, 92, 85, 77],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female']
}
```

```python
# Convert the data to DataFrame
df=pd.DataFrame(data)

df
```

# Frequency Table for Qualitative dta

Frequancy means the number of times a value appears in the data

```python
# frequency table for "category"
category_freq=df["Category"].value_counts()
print("frequency table for 'Category':\n",category_freq)
```

```python
# Code
# @title
plt.figure(figsize=(6, 4))

# Plot for 'Category'

sns.countplot(x='Category', data=df, palette='viridis')
plt.title("Frequency Distribution: Category")
plt.xlabel("Category")
plt.ylabel("Frequency")
```

```python
# frequnecy table for "Gender"
gender_freq=df["Gender"].value_counts()
print("frequency table for 'Gender':\n",gender_freq)
```

```python
# Code
# @title
plt.figure(figsize=(6, 4))

# Plot for 'Category'

sns.countplot(x='Gender', data=df, palette='viridis')
plt.title("Frequency Distribution: Gender")
plt.xlabel("Gender")
plt.ylabel("Frequency")
```

# Frequency table for numerical Data

```python
# frrequency table for "scores"
Scores_freq=df['Scores'].value_counts()
print("frequency table for 'Scores':\n",Scores_freq)
```

```python
# plot for 'score'

sns.countplot(x='Scores',data=df,palette='magma')
plt.title("Frequency Distribution: Score")
plt.xlabel("Scores")
plt.ylabel("Frequency")
```

if the data has many different values it is easier to use intervals of values to present them in a table.

```python
# Creating bins for 'score'

df['Score_bins']=pd.cut(df['Scores'],bins=[70,80,90,100],labels=['70-80','80-90','90-100'])
# frequency table for 'Score_bins'
Score_bins_freq=df['Score_bins'].value_counts().sort_index
print("\nfrequency table for 'score_bins':\n",Score_bins_freq)
```

```python
# Plotting Frequency Distribution for Score
plt.figure(figsize=(6, 4))
sns.countplot(x='Score_bins', data=df, palette='cool', legend=True, hue='Score_bins')
plt.title("Frequency Distribution: Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```
