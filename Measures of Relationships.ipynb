```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Wine Data

```python
from sklearn.datasets import load_wine

wine_data=load_wine(as_frame=True)
wine_df= wine_data.frame
wine_df.head()
```

# Covariance

```python
covariance_matrix=wine_df.cov()
covariance_matrix
```

# Correlation

```python
wine_correlation_matrix=wine_df.corr()
wine_correlation_matrix
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(wine_correlation_matrix, annot=True, cmap="coolwarm")
plt.title('Correlation Heatmap')
plt.show()
```

Correlation: values range from -1 to 1, where values close to 1 indicate a strong positive relationship, close to -1 indicates a strong negative relationship, and close to 0 mean no relationship.



* A high positive correlation (e.g., between alcalinity of ash content and target - quality) may suggest that higher alcalinity of ash rated better.
* Negative covariance might be seen between flavanoids and the target - quality, implying that as one increases, the other decreases.
* Flavanoid and total phenols are highly correlated. so we only need one of them.

# Load_iris

```python
from sklearn.datasets import load_iris

iris_data=load_iris(as_frame=True)
iris_df=iris_data.frame
iris_df
```

```python
iris_correlation_matrix=iris_df.corr()
iris_correlation_matrix
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(iris_correlation_matrix, annot=True,cmap='coolwarm')
plt.title('correlation Heatmap')
plt.show()
```

***Correlation Insights from the Heatmap***:
1. Positive Correlation:
Petal Length and Petal Width: These two features are highly positively correlated (close to +1). This suggests that as petal length increases, petal width also increases.
2. Sepal Length and Petal Length: A moderate positive correlation indicates that longer sepals tend to be associated with longer petals.
Negative Correlation:
3. Sepal Width and Petal Length: There is a slight negative correlation, suggesting that flowers with longer petals tend to have narrower sepals, though this is a weaker relationship.


**Target Variable Correlation:**
1. Petal Dimensions and Target: Petal dimensions (length and width) shows Strong correlation with the target variable indicates that petal dimensions are good predictor of the flower species (class label).
2. Sepal Dimensions and Target: Sepal dimensions (length and width) show relatively weaker correlations with the target, meaning they are less predictive of species compared to petal dimensions.

# Scatter Plot

```python
# Scatter plot between sepal length and sepal width, colored by target (class labels)
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_df['target'], cmap='viridis')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.title('Scatter plot: sepal length vs sepal width (colored by target)')
plt.colorbar(label='Target')
```

```python
# Scatter plot between petal length and petal width, colored by target (class labels)
plt.scatter(iris_df['petal length (cm)'], iris_df['petal width (cm)'], c=iris_df['target'], cmap='viridis')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.title('Scatter plot: petal length vs petal width (colored by target)')
plt.colorbar(label='Target')
```

```python
# Create parwise scatter plot using seaborn's pairplot
sns.pairplot(iris_df, hue='target', markers=["o", "s", "D"])
```

***Scatter Plot Inferences:***


1. Cluster Separation:
* When visualizing Sepal Length vs. Sepal Width and Petal Length vs. Petal Width, the scatter plots reveal clear separation between the three species (color-coded by the target variable).
* Setosa: This species tends to cluster tightly in its own space for both sepal and petal features, indicating that it’s easily distinguishable from the other species.
* Versicolor and Virginica: These two species have overlapping clusters, particularly in the sepal feature space, but they are more distinguishable in the petal feature space (especially in petal length and width).
2. Strong Linear Relationship:
The scatter plots between Petal Length and Petal Width show a clear linear relationship for each species, especially for Versicolor and Virginica. This reinforces the high positive correlation seen in the heatmap.
